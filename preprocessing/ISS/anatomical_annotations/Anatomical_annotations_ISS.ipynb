## Annotating anatomical structures in ISS data

```python
# initialisation 
import os
import panel as pn
import socket
import numpy as np
import tissue_tag as tt 
os.environ["BOKEH_ALLOW_WS_ORIGIN"] = "*"
host = '5011' # set the port to the value in the address bar when operating for instance on farm
```

```python
data_dir = '/nfs/team292/vl6/FetalReproductiveTract/ISS/'
sample = 'CM008'
```

```python
# here you can either read a single image (grayscale or RGB) or generate a virtual H&E from 2 images
 
im,_,ppm = tt.read_image(
    path=data_dir+'CM008.png', # virtual H&E
    contrast_factor=1,
    # background_image_path=path+'Sample_05_THY45_Z5_ch0058.jpg', # panCK
    ppm_image=11.1, # this is round(1/0.09, 1) 
    ppm_out=0.25,
)
```

```python
# define annotations 
annodict = {
    'Unassigned': 'gray',
    'FallopianTube': 'orange', 
    'Uterus': 'red', 
    'WolffianRemnants': 'blue', 
    'Cervix': 'yellowgreen', 
    'UpperVagina': 'forestgreen', 
    'LowerVagina': 'darkcyan', 
    'Clitoris': 'violet', 
    'Penis': 'mediumpurple',
    'LowerTract' : 'darkgrey'
}
labels = np.zeros((im.shape[0],im.shape[1]), dtype=np.uint8) # blank annotation image canvas
```

```python
# use annotator to label tissue regions according to categories indicated above
annotator,render_dict = tt.annotator(im,labels,anno_dict=annodict,use_datashader=True)
pn.io.notebook.show_server(annotator, notebook_url=f'node-14-20:'+host)
```

```python
# This step fills in the shapes created for the pixel classifier

labels = tt.update_annotator(
    imarray=im,
    labels=labels,
    anno_dict=annodict,
    render_dict=render_dict
)
```

```python
isExist = os.path.exists(data_dir+sample+'/'+'tissue_annotations')
if not(isExist):
    os.mkdir(data_dir+sample+'/tissue_annotations/')
    
tt.save_annotation(
    folder=data_dir+sample+'/tissue_annotations/',
    label_image=labels,
    file_name='anatomical_annotations',
    anno_names=list(annodict.keys()),
    anno_colors=list(annodict.values()),
    ppm = ppm
)
```

```python
labels, annotation_order, ppm, annodict = tt.load_annotation(folder=data_dir+sample+'/tissue_annotations/',file_name='anatomical_annotations',load_colors=True)
```

- Load the TissueTag annotated image of the virtual H&E created from an ISS sample
- Translate the pixel level information to a an hexagonal grid space of a set spacing which we can be used to measure the minimal euclidean distances between tissue compartments
- Migrate annotations and measurements to single cell segmentation data by their xy position

```python
# load tissue annotations from TissueTag and migrate annotation to a 15um hexagonal grid and outputs the coordinates in ppm=1 
df_anno = tt.anno_to_grid(
    folder=data_dir+sample+'/tissue_annotations/',
    file_name = 'anatomical_annotations',
    spot_to_spot = 15, # this determines the grid density in microns
    load_colors=True,
)
df_anno['anatomical_annotations'].value_counts()
```

```python
print('calculating distances')
tt.dist2cluster_fast(
    df=df_anno,
    annotation='anatomical_annotations',
    KNN=10
    ) # calculate minimum mean distance of each spot to clusters
```
